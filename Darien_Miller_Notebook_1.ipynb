```
import ssl
import urllib.request
import json 

gcontext = ssl.SSLContext()
apodurlobj = urllib.request.urlopen("https://data.cityofnewyork.us/resource/fhrw-4uyv.json", context=gcontext)
apodread = apodurlobj.read()
data = json.loads(apodread.decode("utf-8"))

list(data)
```
